```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Data Exploration

```python
dataset=pd.read_csv('TelcoCustomerChurn.csv')
dataset
```

```python
df=dataset.drop('customerID', axis=1)
df
```

```python
df.dtypes
```

```python
'''
ok so we found one issue right away.
See how the Monthly Charges is a float
but total is an OBJECT

'''
df.TotalCharges.values
```

```python
'''
its a string so lets change that

# Step 1: Replace spaces and empty strings with NaN
df['TotalCharges'] = df['TotalCharges'].replace([' ', ''], np.nan)

# Step 2: Convert the column to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

# Step 3: Fill NaN values with 0 (optional)
df['TotalCharges'] = df['TotalCharges'].fillna(0)

The above code overdid it really
Because I forgot the values were just " " . Now total charges are 0 
So we cannot do that
'''
df.iloc[488]
```

```python
'''
see how there are no total charges
I will remove them now

'''
df[df.TotalCharges==' '].shape #this is like (11,20), ie 11 values. I will remove it
df1=df[df.TotalCharges!=' ']
df1.shape
```

```python
'''now i will use the command'''
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
df1.TotalCharges.dtypes #it is float now.
```

```python
# I will make histogram of people who are leaving to ones who are not, by judging the tenure
tenure_churn_no=df1[df1.Churn=='No'].tenure
tenure_churn_yes=df1[df1.Churn=='Yes'].tenure


plt.hist([tenure_churn_yes,tenure_churn_no], label=['ChurnYes', 'ChurnNo'])
plt.xlabel("Tenure (in months)")
plt.ylabel("Number of Customers")

plt.legend()
```

```python
'''this chart told us about how the Tenure affected the churn. People with high tenure are not leaving'''
'''lets do the same with others'''
# I will make histogram of people who are leaving to ones who are not, by judging the tenure
MonthlyCharges_churn_no=df1[df1.Churn=='No'].MonthlyCharges
MonthlyCharges_churn_yes=df1[df1.Churn=='Yes'].MonthlyCharges


plt.hist([MonthlyCharges_churn_yes,MonthlyCharges_churn_no], label=['ChurnYes', 'ChurnNo'])
plt.xlabel("MonthlyCharges (in Currency)")
plt.ylabel("Number of Customers")

plt.legend()
```
